## Attempt 2: Using librosa and mp3 Files

```python
import librosa
# librosa is not compatible with numpy 2
# Check if I am using a version of numpy < 2
import numpy as np
print(np.__version__)
```

```python
# Load the audio file
data, sample_rate = librosa.load('ron-minis-cut.mp3', sr=None)
```

```python
# data now contains the amplitude values, sample_rate is the sampling rate of the audio
print("Sample Rate:", sample_rate)
print("Data Shape (number of signals):", data.shape)
print("Number of signals/sample rate = audio duration =", round(len(data)/sample_rate, 2), "seconds")
print("Amplitude Values:", data)
```

librosa returns the audio as floating-point values, thus the amplitude values are within the \[-1.0, 1.0\] range. Now I would like to visualize the data.

```python
import matplotlib.pyplot as plt
import librosa.display
```

```python
plt.figure().set_figwidth(12)
librosa.display.waveshow(data, sr=sample_rate)
```
